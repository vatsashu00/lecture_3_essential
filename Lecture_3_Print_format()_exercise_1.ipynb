# Role of Underscore(_) in Python

# You will find max six different uses of underscore(_).

```python
5 + 4
```

```python
_ 
```

```python
_ + 6
```

```python
a = _
```

```python
## ignoring a value
a, _, b = (1, 2, 3) # a = 1, b = 3
print(a, b)
```

```python
## ignoring multiple values
## *(variable) used to assign multiple value to a variable as list while unpacking
## it's called "Extended Unpacking", only available in Python 3.x
a, *_, b = (7, 6, 5, 4, 3, 2, 1)
print(a, b)
```

# Separating Digits Of Numbers

```python
million = 1_000_000
binary = 0b_0010
octa = 0o_64
hexa = 0x_23_ab

print(million)
print(binary)
print(octa)
print(hexa)
```
